```python
import pandas as pd
import requests

# API endpoint URL
api_url = "https://imdb-api.com/API/AdvancedSearch/k_ubtu0kka?title_type=feature&groups=top_250&languages=en&count=250"

# Make a GET request to the API
response = requests.get(api_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data from the response
    data = response.json()

    # Extract the "results" key from the JSON data
    movie_list = data["results"]

    # Convert the list of movies to a pandas DataFrame
    movie_df = pd.DataFrame(movie_list)

    # Display the first few rows of the DataFrame
    print(movie_df.head())
else:
    print("Failed to fetch data from IMDb API.")
```
