```python
import pandas as pd 
import pathlib as path
```

```python
data = pd.read_csv('combined_box_office.csv')
data
```

```python
data_sorted = data.sort_values(by='year_x',ascending=False)
data_sorted
```

```python
#filtered and exclude years 
exclude_years = [2020,2021,2022,2023]

filtered_year = data[~data['year_x'].isin(exclude_years)]

filtered_year
```

```python
clean_df = filtered_year[["id", "rank_x","title_x","worldwideLifetimeGross_x","domesticLifetimeGross_x","domestic_x","foreignLifetimeGross_x","foreign_x","year_x"]]
clean_df
```

```python
column_name_map = {
    'id':'ID',
    'rank_x':'rank',
    'title_x':'title',
    'worldwideLifetimeGross_x':'Worlwide revenue',
    'domesticLifetimeGross_x': 'Domestic revenue',
    'domestic_x': 'Domestic Percentage',
    'foreignLifetimeGross_x': 'Foreign revenue',
    'foreign_x': 'Foreign Percentage',
    'year_x': 'Year'
}

clean_df.rename(columns=column_name_map, inplace=True)

clean_df
```

```python
clean_df.to_csv ('cleaned_box_office.csv', index=False)
```

```python
clean_sorted = clean_df.sort_values(by='rank',ascending=True)
clean_sorted
```
