```python
import pandas as pd
import pathlib as path
```

```python
#read csv files 
file1 ="box_office_all_time.csv"
file2 ="box_office.csv"

box_office_all_time_df = pd.read_csv(file1)
box_office_all_time_df

box_office_df = pd.read_csv(file2)
box_office_df
```

```python
#combined the 2 dataframes
combined_df = pd.merge(box_office_all_time_df,box_office_df, on="id")
combined_df.head(20)
```

```python
# export datafram to csv files 
combined_df.to_csv ('combined_box_office.csv', index=False)
```
